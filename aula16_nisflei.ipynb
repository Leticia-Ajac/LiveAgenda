```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

#gerando base de dados
x,y = make_blobs(n_samples=1200, centers=3, cluster_std=0.70, random_state=0)

#print(x)
```

```python
#executar o cluster

k_means = KMeans(n_clusters=3, init='k-means++', max_iter=500, n_init=10, random_state=0)

pred = k_means.fit_predict(x)
```

```python
#formatação gráfico
plt.figure(figsize=(20,8))
plt.scatter(x[:,0], x[:,1], c=k_means.labels_)

#centroides
plt.scatter(k_means.cluster_centers_[:,0], k_means.cluster_centers_[:,1], s=200, c='orange')

plt.show
```

```python
#cotovelo

wcss=[]
for i in range(1, 10):
  kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=300, random_state=0)
  pred = kmeans.fit(x)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,10), wcss)
plt.title('Metodo do Cotovelo')
plt.show
```
